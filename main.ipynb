```python
from transformers import pipeline
def main():
    print("Welcome to the text summarizer.Please enter your text")
    summarizer=pipeline("summarization",model="facebook/bart-large-cnn")
    while True:
        user_input=input("Enter the text here:")
        if user_input.lower()=="exit":
            print("Exiting the program")
            break
        if len(user_input.split())<20:
            print("The input is too short to summarize")
            continue
        print("Summarizing...")
        summary=summarizer(user_input,max_length=100,min_length=30,do_sample=False)[0]["summary_text"]
        print("\nSummarized text")
        print(summary)
if __name__=="__main__":
    main()
```
